```python
Data = pd.DataFrame()
def preprocessing(text):
    text_without_emoji = emoji.get_emoji_regexp().sub(r'', text)
    text_without_digits = re.sub(r'\d+', '', text_without_emoji)
    text_without_alnum = re.sub(r'\W+', ' ', text_without_digits)
    tokens = word_tokenize(text_without_alnum)
    tokens_without_punct = [i for i in tokens if i not in punctuations]
    low_tokens = [i.lower() for i in tokens_without_punct]
    words_without_stop = [i for i in low_tokens if i not in stopwords]
    lemms = [morph.parse(i)[0].normal_form for i in words_without_stop]
    return ' '.join(lemms)
```

```python
for index, row in Data.iterrows():
    processed_text = preprocessing (row['text'])
    print(processed_text)
```

# Решение задачи классификации

Возьмем выбранный ранее список векторов из предыдущего пункта.
Разделим данные на обучающее и тестовое множество.
Выберем метрику для релевантности поста - количество лайков. Сгруппируем все данные в несколько классов (популярный/непопулярный пост).
Обучим несколько классификаторов и оценим их работу с помощью метрик. Далее выберем наилучший классификатор.

В качестве метрики популярности поста выбираем - количество лайков, где популярный пост - тот, у которого количество лайков больше медианного значения, а непопулярный - у кого количество лайков соответсвует медианному значению и ниже.

```python
# Находим медиану по колонке 'likes'
median_likes = df['likes'].median()

# результат
print("Медиана по колонке 'likes':", median_likes)
```

```python
def likes_to_binary(likes):
    if likes > median_likes:
        return 1
    else:
        return 0

# Применяем функцию к столбцу 'likes' с помощью метода apply
df['target'] = df['likes'].apply(lambda x: likes_to_binary(x))
```

# Подбор гиперпараметров

Осуществить подбор гиперпараметров удалось не совсем в полной мере. Это связано первоочередно с нехваткой памяти или других ресурсов системы. Поэтому в этом коде я уменьшила количество параметров для перебора, чтобы снизить нагрузку на систему. Также я использовала только 3-кратную кросс-валидацию (cv=3) и ограничила число параллельно запущенных процессов (n_jobs=1).

Определение параметров для Grid Search для SVM

```python
from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
param_grid_svm = {
    'C': [0.1, 0.5, 1.0],
    'max_iter': [1000, 2000]
}

# Инициализация и обучение Grid Search для SVM
grid_search_svm = GridSearchCV(LinearSVC(penalty='l2', class_weight='balanced', random_state=42), param_grid_svm, cv=3, scoring='f1', n_jobs=1, verbose=1)
grid_search_svm.fit(X_train, y_train)
best_model_svm = grid_search_svm.best_estimator_
print("Лучшие гиперпараметры для SVM:", grid_search_svm.best_params_)
svm_predictions = best_model_svm.predict(X_test)
svm_f1_score = f1_score(y_test, svm_predictions)
print(f"F1-score на тестовой выборке для SVM: {svm_f1_score}")
```

В ходе подбора гиперпараметров к модели SVM получили результат f1_score, превышающий предыдущий - 0,658.  

Определение параметров для Grid Search для KNN

```python
# Определение параметров для Grid Search для KNN
param_grid_knn = {
    'n_neighbors': [3, 5, 7]
}

# Инициализация и обучение Grid Search для KNN
grid_search_knn = GridSearchCV(KNeighborsClassifier(), param_grid_knn, cv=3, scoring='f1', n_jobs=1, verbose=1)
grid_search_knn.fit(X_train, y_train)
best_model_knn = grid_search_knn.best_estimator_
print("Лучшие гиперпараметры для KNN:", grid_search_knn.best_params_)
knn_predictions = best_model_knn.predict(X_test)
knn_f1_score = f1_score(y_test, knn_predictions)
print(f"F1-score на тестовой выборке для KNN: {knn_f1_score}")
```

Результат f1_score при подборе гиперпараметров для KNN - 0,625 - оказался выше своего результата в предыдущем пункте, но модель по качеству не смогла обогнать SVM.

Определение параметров для Grid Search для Decision Tree

```python
param_grid_dt = {
    'max_depth': [None, 10, 20, 30]
}
grid_search_dt = GridSearchCV(DecisionTreeClassifier(random_state=42), param_grid_dt, cv=3, scoring='f1', n_jobs=1, verbose=1)
grid_search_dt.fit(X_train, y_train)
best_model_dt = grid_search_dt.best_estimator_
print("Лучшие гиперпараметры для Decision Tree:", grid_search_dt.best_params_)
dt_predictions = best_model_dt.predict(X_test)
dt_f1_score = f1_score(y_test, dt_predictions)
print(f"F1-score на тестовой выборке для Decision Tree: {dt_f1_score}")
```

Подбором гиперпараметров значительно улучшилась модель Decision Tree и показала лучший из приведенных f1_score - **0,659**. Ее будем использовать для тестирования модели на новом посте далее.
